```python
def sub_cuisine(x):
    pd.set_option('display.max_colwidth', 100)
    subcuisine = ', '.join(us_restaurants.loc[us_restaurants.cuisine == x].categories.astype(str))
    cats=pd.DataFrame(subcuisine.split(','),columns=['category'])
    x=cats.category.value_counts()
    return pd.DataFrame(x)  
```

```python
def word_cloud(x):
    from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
    # Start with one review:
    text =', '.join(us_restaurants.loc[us_restaurants.cuisine == x].categories.astype(str))


    # Create and generate a word cloud image:
    stop_words = ['Restaurants', 'American Traditional', 'American', 'New','American New','Traditional', 'Japanese'] + list(STOPWORDS)

    wordcloud = WordCloud(
        stopwords = stop_words
    ).generate(text)

    # Display the generated image:
    fig = plt.gcf()
    fig.set_size_inches(13, 6)

    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    #fig.savefig('test2png.png', dpi=100)
    return
```
